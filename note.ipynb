```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('r_DataQR.csv')

df.head()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Supposons que vous ayez déjà chargé votre dataframe df et effectué les étapes nécessaires pour le prétraitement des données.

vectorizer = CountVectorizer()
df['Combined_text'] = df['Question'] + ' ' + df['Réponse correcte']
X = vectorizer.fit_transform(df['Combined_text'])
answers = df['Réponse correcte'].values

# Affichage de la question et saisie utilisateur
question_idx = int(input("Entrez le numéro de la question: "))
question = df.iloc[question_idx]['Question']
user_input = input(f"{question}\nVotre réponse : ")

# Transformation de la saisie utilisateur en vecteur
user_input_vector = vectorizer.transform([user_input])

# Calcul de la similarité cosinus entre la saisie utilisateur et les réponses du dataset
cos_similarities = cosine_similarity(user_input_vector, X)

# Seuil de similarité
similarity_threshold = 0.7  # Ajustez ce seuil selon vos besoins

# Trouver la réponse la plus similaire
most_similar_idx = cos_similarities.argmax()
max_similarity = cos_similarities[0, most_similar_idx]

if max_similarity >= similarity_threshold:
    most_similar_answer = answers[most_similar_idx]
    print("Bonne réponse ! La réponse correcte est :", most_similar_answer)
else:
    print("Mauvaise réponse. ")
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Assuming you have already loaded your dataframe df and performed necessary data preprocessing.

# Combine question and answer into a single text
df['Combined_text'] = df['Question'] + ' ' + df['Réponse correcte']

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['Combined_text'])
answers = df['Réponse correcte'].values

# Precompute cosine similarities
cos_similarities = cosine_similarity(X)

# Display the question and get user input
try:
    question_idx = int(input("Entrez le numéro de la question: "))
    question = df.iloc[question_idx]['Question']
    user_input = input(f"{question}\nVotre réponse : ")

    # Transform user input into a vector (if vectorizer parameters change, transform here)
    user_input_vector = vectorizer.transform([user_input])

    # Calculate cosine similarity between user input and dataset answers using sparse matrix operations
    user_cos_similarities = cosine_similarity(user_input_vector, X, dense_output=False).toarray()[0]

    # Set similarity threshold
    similarity_threshold = 0.7  # Adjust as needed

    # Find the most similar answer using np.argmax for efficiency
    most_similar_idx = np.argmax(user_cos_similarities)
    max_similarity = user_cos_similarities[most_similar_idx]

    # Print debugging outputs
    print("Cosine Similarities:", user_cos_similarities)
    print("Most Similar Index:", most_similar_idx)
    print("Max Similarity:", max_similarity)

    # Check if the max similarity is above the threshold
    if max_similarity >= similarity_threshold:
        most_similar_answer = answers[most_similar_idx]
        print("Bonne réponse ! La réponse correcte est :", most_similar_answer)
    else:
        print("Mauvaise réponse.")
except ValueError:
    print("Erreur : Veuillez entrer un numéro de question valide.")
```
